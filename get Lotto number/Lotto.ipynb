```python
import pandas as pd

class Lotto :
    Lotto_number = pd.DataFrame()
            
    def __init__(self, Lotto_number):
        self.Lotto_number = Lotto_number
    
    def main(self):
        self.GetLottoNumber()
        
    
    def GetLottoNumber(self):
        for i in self.Lotto_number.columns:
            print(int(self.Lotto_number[i].argmax()) + 1, end=" ")
            
Lotto_number = pd.read_excel("Lotto_number_count.xlsx")

Lotto = Lotto(Lotto_number)

Lotto.main()
```
