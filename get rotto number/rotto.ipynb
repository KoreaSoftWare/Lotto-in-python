```python
import pandas as pd

class Rotto :
    rotto_number = pd.DataFrame()
            
    def __init__(self, rotto_number):
        self.rotto_number = rotto_number
    
    def main(self):
        self.GetRottoNumber()
        
    
    def GetRottoNumber(self):
        for i in self.rotto_number.columns:
            print(int(self.rotto_number[i].argmax()) + 1, end=" ")
            
rotto_number = pd.read_excel("rotto_number_count.xlsx")

rotto = Rotto(rotto_number)

rotto.main()
```
